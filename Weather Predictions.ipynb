### Format the data


```python
# Create a data set in the format of...
# [[[1], [2], [3], [4], [5]]] [date6, 6]
# [[[2], [3], [4], [5], [6]]] [date7, 7]
# [[[3], [4], [5], [6], [7]]] [date8, 8]

def df_to_X_y(df, window_size=5):
    df_as_np = df.to_numpy()
    X = []
    y = []
    dates = []
    
    for i in range(len(df_as_np)-window_size):
        row = [a for a in df_as_np[i:i+window_size]]
        X.append(row)

        date = df.index[i]
        label = [date, df_as_np[i+window_size][0]] # We capture the date of the prediction  
        y.append(label)     

    return np.array(X), np.array(y)
```

```python
# Set the previous n days to 5
WINDOW_SIZE = 5

X1, y1 = df_to_X_y(temp, WINDOW_SIZE)
X1.shape, y1.shape
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import * # Yes, I am lazy :)
from tensorflow.keras.callbacks import ModelCheckpoint
from tensorflow.keras.losses import MeanSquaredError
from tensorflow.keras.metrics import RootMeanSquaredError
from tensorflow.keras.optimizers import Adam
```

```python
model1 = Sequential()
model1.add(InputLayer((WINDOW_SIZE, 1)))
model1.add(LSTM(64))
model1.add(Dense(8, 'relu'))
model1.add(Dense(1, 'linear'))

model1.summary()
```

```python
# Start a new data set based on temp and time
temp_df = pd.DataFrame(temp)
temp_df['Seconds'] = temp_df.index.map(pd.Timestamp.timestamp)
temp_df
```

```python
# Temporal Sin and Cos analysis
day = 60*60*24 #sec/day = sec/min x min/hr x hr/day
year = 365.2425*day #sec/yr = sec/day x day/yr
```

```python
temp_df['Day sin'] = np.sin(temp_df['Seconds'] * (2* np.pi / day))
temp_df['Day cos'] = np.cos(temp_df['Seconds'] * (2 * np.pi / day))
temp_df['Year sin'] = np.sin(temp_df['Seconds'] * (2 * np.pi / year))
temp_df['Year cos'] = np.cos(temp_df['Seconds'] * (2 * np.pi / year))
temp_df.head()
```

```python
# Drop the seconds column
temp_df = temp_df.drop('Seconds', axis=1)
temp_df.head()
```

```python
temp_df.to_numpy()
```

```python
# Create a data set in the format of...
# [[[X1_1 X2_1... XN_1], [X1_2 X2_2... XN_2],... [Size of window-1]]] [y_WindowSize]
# [[[X1_2 X2_2... XN_2], [X1_3 X2_3... XN_3],... Size of window-1]] [ y_WindowSize+1]

def df_to_X_y4(df, window_size=5):
    df_as_np = df.to_numpy()
    X = []
    y = []
    dates = []
    
    for i in range(len(df_as_np)-window_size):
        row = [a for a in df_as_np[i:i+window_size]]
        X.append(row)

        date = df.index[i]
        label = [date, df_as_np[i+window_size][0]]
        y.append(label)     

    return np.array(X), np.array(y)
```

```python
def preprocess5(X):
    Xp = X.copy()
    Xp[:, :, 0] = (X[:, :, 0] - p_training_mean5)/p_training_std5
    Xp[:, :, 1] = (X[:, :, 1] - temp_training_mean5)/temp_training_std5
    return Xp

def preprocess_output5(y):
    yp = y.copy()
    yp[:, 1] = (y[:, 1] - p_training_mean5)/p_training_std5
    yp[:, 2] = (y[:, 2] - temp_training_mean5)/temp_training_std5
    return yp
```
