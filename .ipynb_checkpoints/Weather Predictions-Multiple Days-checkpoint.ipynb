## Predict the next m day's of temperature based on previous n days of temperature

### Format the data


```python
# window_size = n days of previous temperatures
# prediction_size = m days of future temperature
# Create a data set in the format of...
# [[[1], [2], [3], [4], [5]]] [date6, 6 7 8]
# [[[2], [3], [4], [5], [6]]] [date7, 7 8 9]
# [[[3], [4], [5], [6], [7]]] [date8, 8 9 10]

def df_to_X_y(df, window_size=10, prediction_size=5):
    X = []
    y = []
    dates = []
    for i in range(df.shape[0]-window_size-prediction_size):
        row = df.iloc[i:i+window_size,0].values.flatten().tolist()
        X.append(row)

        date = df.index[i+window_size]
        
        yrow = df.iloc[i+window_size:i+window_size+prediction_size,:].values.flatten().tolist()
        label = [date, yrow ] # We capture the date of the prediction
        y.append(label)     

    return np.array(X), np.array(y)
```

```python
# Set the previous n days to 5
WINDOW_SIZE = 10
PREDICTION_SIZE = 5

X1, y1 = df_to_X_y(temp, WINDOW_SIZE, PREDICTION_SIZE)
X1.shape, y1.shape
```

```python
y1
```

### Create datasets

```python
from sklearn.model_selection import train_test_split
```

```python
# 80% training data used to create model
# 10% val data used in model validation
# 10% test datanever seen during model development

train_end1 = int(len(X1)*0.8)
test_end1 = int(len(X1)*0.9)

X_train1, y_train1 = X1[:train_end1], y1[:train_end1]
X_val1, y_val1 = X1[train_end1:test_end1], y1[train_end1:test_end1]
X_test1, y_test1 = X1[test_end1:], y1[test_end1:]
```

```python
X_train1.shape, y_train1.shape, X_val1.shape, y_val1.shape, X_test1.shape, y_test1.shape
```

### Create LSTM Model

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import * # Yes, I am lazy :)
from tensorflow.keras.callbacks import ModelCheckpoint
from tensorflow.keras.losses import MeanSquaredError
from tensorflow.keras.metrics import RootMeanSquaredError
from tensorflow.keras.optimizers import Adam
```

```python
model1 = Sequential()
model1.add(InputLayer((WINDOW_SIZE, 1)))
model1.add(LSTM(64, return_sequences=True))
model1.add(LSTM(32))
model1.add(Dense(16, 'relu'))
model1.add(Dense(PREDICTION_SIZE, 'linear'))

model1.summary()
```

```python
# Use ModelCheckpoint to save the best model
cp1 = ModelCheckpoint('model1_lstm_5d/', save_best_only=True)

# Create model
model1.compile(loss=MeanSquaredError(), 
               optimizer=Adam(learning_rate=0.0001), 
               metrics=[RootMeanSquaredError()])
```

```python
# Fit the model
y_train1_d = np.array(y_train1[:,1].tolist())
y_val1_d = np.array(y_val1[:,1].tolist())
y_test1_d = np.array(y_test1[:,1].tolist())
```

```python
# Don't need to necessarly normalize or scale the data (univariate)
history = model1.fit(X_train1, 
           y_train1_d, 
           validation_data=(X_val1, y_val1_d), 
           epochs=15, 
           callbacks=[cp1])
```

# Add new variables: Sin Cos Temporal Analysis

https://towardsdatascience.com/cyclical-features-encoding-its-about-time-ce23581845ca

```python
# a quick plot to see what the data looks like
# nice cycles... need to try a cos sin analysis

temp = dataset[['T (degC)']]
temp
```

```python
# Start a new data set based on temp and time
temp_df = pd.DataFrame(temp)
temp_df['Seconds'] = temp_df.index.map(pd.Timestamp.timestamp)
temp_df
```

```python
# Temporal Sin and Cos analysis
day = 60*60*24 #sec/day = sec/min x min/hr x hr/day
year = 365.2425*day #sec/yr = sec/day x day/yr
```

```python
temp_df['Day sin'] = np.sin(temp_df['Seconds'] * (2* np.pi / day))
temp_df['Day cos'] = np.cos(temp_df['Seconds'] * (2 * np.pi / day))
temp_df['Year sin'] = np.sin(temp_df['Seconds'] * (2 * np.pi / year))
temp_df['Year cos'] = np.cos(temp_df['Seconds'] * (2 * np.pi / year))
temp_df.head()
```

```python
# Drop the seconds column
temp_df = temp_df.drop('Seconds', axis=1)
temp_df.head()
```

```python
temp_df.to_numpy()
```

```python
# Create a data set in the format of...
# [[[X1_1 X2_1... XN_1], [X1_2 X2_2... XN_2],... [Size of window-1]]] [y_WindowSize]
# [[[X1_2 X2_2... XN_2], [X1_3 X2_3... XN_3],... Size of window-1]] [ y_WindowSize+1]

def df_to_X_y4(df, window_size=5):
    df_as_np = df.to_numpy()
    X = []
    y = []
    dates = []
    
    for i in range(len(df_as_np)-window_size):
        row = [a for a in df_as_np[i:i+window_size]]
        X.append(row)

        date = df.index[i]
        label = [date, df_as_np[i+window_size][0]]
        y.append(label)     

    return np.array(X), np.array(y)
```

```python
WINDOW_SIZE = 6 #So we can see how the shape changes

X4, y4 = df_to_X_y4(temp_df, WINDOW_SIZE)
X4.shape, y4.shape
```

```python
temp_df.tail()
```

```python
# Setup data
train_end = int(len(X4)*0.8)
test_end = int(len(X4)*0.9)

X_train4, y_train4 = X4[:train_end], y4[:train_end]
X_val4, y_val4 = X4[train_end:test_end], y4[train_end:test_end]
X_test4, y_test4 = X4[test_end:], y4[test_end:]
```

```python
p_temp_df = pd.concat([dataset['p (mbar)'], temp_df], axis=1)
p_temp_df.head()
```

```python
# Create a data set in the format of...
# [[[X1_1 X2_1... XN_1], [X1_2 X2_2... XN_2],... [Size of window-1]]] [y1_WindowSize y2_WindowSize ]
# [[[X1_2 X2_2... XN_2], [X1_3 X2_3... XN_3],... Size of window-1]]   [y1_WindowSize+1 y1_WindowSize+1] ]

def df_to_X_y5(df, window_size=7):
    df_as_np = df.to_numpy()
    X = []
    y = []
    dates = []
    
    for i in range(len(df_as_np)-window_size):
        row = [a for a in df_as_np[i:i+window_size]]
        X.append(row)

        date = df.index[i]
        label = [date, df_as_np[i+window_size][0], df_as_np[i+window_size][1]]
        y.append(label)     

    return np.array(X), np.array(y)
```

```python
WINDOW_SIZE5 = 7 #Again, just changing the size to see the impact on shape

X5, y5 = df_to_X_y5(p_temp_df, WINDOW_SIZE5)
X5.shape, y5.shape
```

```python
# Setup data

train_end = int(len(X5)*0.8)
test_end = int(len(X5)*0.9)

X_train5, y_train5 = X5[:train_end], y5[:train_end]
X_val5, y_val5 = X5[train_end:test_end], y5[train_end:test_end]
X_test5, y_test5 = X5[test_end:], y5[test_end:]

X_train5.shape, y_train5.shape, X_val5.shape, y_val5.shape, X_test5.shape, y_test5.shape
```

```python
# Normalize pressure and temperature

p_training_mean5 = np.mean(X_train5[:, :, 0])
p_training_std5 = np.std(X_train5[:, :, 0])

temp_training_mean5 = np.mean(X_train5[:, :, 1])
temp_training_std5 = np.std(X_train5[:, :, 1])
```

```python
def preprocess5(X):
    Xp = X.copy()
    Xp[:, :, 0] = (X[:, :, 0] - p_training_mean5)/p_training_std5
    Xp[:, :, 1] = (X[:, :, 1] - temp_training_mean5)/temp_training_std5
    return Xp

def preprocess_output5(y):
    yp = y.copy()
    yp[:, 1] = (y[:, 1] - p_training_mean5)/p_training_std5
    yp[:, 2] = (y[:, 2] - temp_training_mean5)/temp_training_std5
    return yp
```

```python
def plot_predictions2(model, X, y, start=0, end=100):
  predictions = model.predict(X)
  p_preds, temp_preds = predictions[:, 0], predictions[:, 1]
  p_actuals, temp_actuals = y[:, 0], y[:, 1]
  df = pd.DataFrame(data={'Temperature Predictions': temp_preds,
                          'Temperature Actuals':temp_actuals,
                          'Pressure Predictions': p_preds,
                          'Pressure Actuals': p_actuals
                          })
  plt.plot(df['Temperature Predictions'][start:end])
  plt.plot(df['Temperature Actuals'][start:end])
  plt.plot(df['Pressure Predictions'][start:end])
  plt.plot(df['Pressure Actuals'][start:end])
  return df[start:end]
```

```python
# Plot the results, without postprocessing
start, end = 0, 100
post_processed_df = plot_predictions2(model5, X_test5p, y_test5_d, start, end)
post_processed_df
```

```python
# Plot postprocessed temp

plt.xticks(rotation=45)
plt.plot(y_test5p[:,0][start:end], postprocess_temp(post_processed_df['Temperature Predictions'][start:end]))
plt.plot(y_test5p[:,0][start:end], postprocess_temp(post_processed_df['Temperature Actuals'][start:end]))
plt.legend(['Temperature Predictions', 
            'Temperature Actuals'])
plt.title("model5_p-t_lstm: temperature")
```

```python
# Plot postprocessed pressure

plt.xticks(rotation=45)
plt.plot(y_test5p[:,0][start:end],post_processed_df['Pressure Predictions'][start:end])
plt.plot(y_test5p[:,0][start:end],post_processed_df['Pressure Actuals'][start:end])
plt.legend(['Pressure Predictions', 
            'Pressure Actuals'])
plt.title("model5_p-t_lstm: pressure")
```
